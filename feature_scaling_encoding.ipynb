# Feature scaling and Feature encoding

# 1-Feature Scaling
we can do feature scaling on numeric dataset \
# types
- Minmax scaling
- standard scaling | z score normalization
- Roubust scaling
- logrithmic scaling



```python
# import important libraries
import sklearn
import pandas as pd
import seaborn as sns
```

```python
df= sns.load_dataset('titanic')
df.head()
```

# 1-Min Max scalling

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
df['fare_scaled']=scaler.fit_transform(df[['fare']])
df['fare_scaled'].head()
```

```python
df.head()
```

# 2-standard scaling or z-score normalization

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
df['age_scaled']=scaler.fit_transform(df[['age']])
df['age_scaled'].head()
```

# 3-Robust scaling

```python
from sklearn.preprocessing import RobustScaler
scaler= RobustScaler()
df['age_scaled']=scaler.fit_transform(df[['age']])
df['age_scaled'].head()
```

# 4. Logrithmic scaling/ Normalization

```python
import numpy as np
import pandas as pd

#random data with outliers
data = {'numbers': [10000, 20000, 30000, 1000000, 50000]}
df = pd.DataFrame(data)

# Log Transform
df['numbers_log'] = np.log(df['numbers'])
df['numbers_log2'] = np.log2(df['numbers'])
df['numbers_log10'] = np.log10(df['numbers'])
df.head()
```

# 2- Feature Encoding
we do feature encoding on categorical data set
- one hot encoding
- label encoding
- ordinal encoding

# one hot encoding

```python
df=sns.load_dataset('iris')
df.head()
```

```python
# One-Hot Encoding
# it make 3 different columns on the base of species true value represent 1 and false with 0
encoded_data = pd.get_dummies(df, columns=['species'])
print(encoded_data)
```

# 2-label encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
df['encoded_species'] = encoder.fit_transform(df['species'])
df.sample(20)
```

# 3-ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
encoder= OrdinalEncoder(categories=[['versicolor','virginica','setosa']]  )#versicolor=0','virginica=1','setosa=2
df['encoded_species'] = encoder.fit_transform(df[['species']])
df.sample(20)
```
